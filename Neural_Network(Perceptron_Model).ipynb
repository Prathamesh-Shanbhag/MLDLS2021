# ***Perceptron Model***
* A single-layer perceptron is the basic unit of a neural network. A 
perceptron consists of input values, weights and a bias, a weighted sum and activation function.

```python
#importing All Required Libraries
import tensorflow as tf
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
# Reading from CSV Dataset file.
dataset = pd.read_csv('CO2_dataog.csv')

print(dataset)
```

```python
# Assigning the values to two variables X & Y using List Slicing
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:,-1].values

print(X)
print(Y)
```

```python
# Defining a single layer/ single Neuron Model
model = tf.keras.Sequential([keras.layers.Dense(units=3, activation='relu', input_shape=[1])])
model.add(Dense(1))
print(model.summary())
```

```python
tf.keras.utils.plot_model(
    model,
    to_file="model.png",
    show_shapes=True,
    show_dtype=True,
    show_layer_names=True,
    rankdir="TB",
    expand_nested=False,
    dpi=96,
)
```

```python
# Defining the model evaluation for backpropogation
model.compile(optimizer='sgd', loss='mean_squared_error')
```

* Backpropagation, short for "backward propagation of errors," is an algorithm for supervised learning of artificial neural networks using gradient descent. Here we are using sgd as the optimizer.
* SGD-Stochastic Gradient Descent (SGD) is a simple yet very efficient approach to fitting. It find the model parameters that correspond to the best fit between predicted and actual outputs. 
* In an artificial neural network and an error function, the method calculates the gradient of the error function with respect to the neural network's weights.
* Loss is calculated as mean squared error.(Which should reduce with each epoch)

```python
# Training the model in cycles or iterations both forwards & backwards on the Neural Network.
run_model = model.fit(X,Y, epochs=500)
```

Loss here is high because data in X & Y is not in relation: y =(x/2) - 4

```python
print(model.predict([112.0]))
```

```python
# List all data present in history(All The Intermediate Values Gained during each iteration.)
print(run_model.history.keys())
```

```python
# Visualizing History for Accuracy
plt.plot(run_model.history['loss'])
plt.title('Model Loss v/s Epochs')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show()
```
