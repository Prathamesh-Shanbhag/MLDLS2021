```python
# Importing the Required Libraries Numpy,Pandas & MatplotLib.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading from CSV Dataset file.
dataset = pd.read_csv('CO2_data.csv')

print(dataset)
```

```python
# Assigning the values to two variables X & Y using List Slicing
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:,-1].values

print(X)
print(Y)
```

```python
# Visualizing the data using matplotlib.pyplot On a 2D Graph
plt.xlabel('Fuel Consumption(L/100KM)')
plt.ylabel('C02 Emissions(g/km)')
plt.title('Fuel Consumption v/s CO2 Emissions')
plt.scatter(X,Y,color='blue',marker='*')
```

```python
#Splitting the data into Training and Testing Data 
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2)
print(X_train.shape, X_test.shape)
```

```python
#Fitting Simple Linear Regression Model using X_train & Y_train values
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
#Predicting the test results
Y_pred = model.predict(X_test)
print(Y_pred)
print(Y_train)
```

```python
#Calculate Accuracy
accuracy = model.score(X_test,Y_test)
print(accuracy)
```

```python
plt.scatter(X,Y,color='blue',marker='*')
plt.xlabel('Fuel Consumption(L/100KM)')
plt.ylabel('C02 Emissions(g/km)')
plt.title('Fuel Consumption v/s CO2 Emissions')
plt.plot(X_train, model.predict(X_train), color='red')
plt.show()

plt.scatter(X_test,Y_test,color='red',marker='+')
plt.xlabel('Fuel Consumption(L/100KM)')
plt.ylabel('C02 Emissions(g/km)')
plt.title('Fuel Consumption v/s CO2 Emissions')
plt.plot(X_test, model.predict(X_test), color='blue')
plt.show()
```

```python
# Predicting on an external Data
test_data = np.array([[112]])
test_pred = model.predict(test_data)
print(test_pred)
```

```python
fuelConsumption = int(input("What is your fuel consumption in L/100km? "))
co2Emission = model.predict(np.array([[fuelConsumption]])) 
print('Estimated CO2 Emission with' + str(fuelConsumption) + 'Litres of Fuel is ' + str(co2Emission[0]) + ' gm per kilometer')
```
